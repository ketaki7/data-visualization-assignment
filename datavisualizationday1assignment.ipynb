```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#assignment no 1
#plot a line plot between a and b:
x=np.arange(40,50)
y=np.arange(50,60)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,color="black",marker="o", linewidth=1, markersize=6)
```

```python
#assignment 2
#Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

plt.xlabel("Days")
plt.ylabel("sales 1")
plt.plot(days,sales_1,color="black",marker="o", linewidth=1, markersize=6)
```

```python
plt.xlabel("Days")
plt.ylabel("sales 2")
plt.plot(days,sales_2,color="black",marker="o", linewidth=1, markersize=6)
```

```python
#assignment no 3

x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplot(3,3,1)
plt.plot(x,y1,'r--')

plt.subplot(3,3,2)
plt.plot(x,y2,'bo--')

plt.subplot(3,3,3)
plt.plot(x,y3,'g*--')

plt.subplot(3,3,4)
plt.plot(x,y4,'b-')

plt.subplot(3,3,5)
plt.plot(x,y5,'go-')

plt.subplot(3,3,6)
plt.plot(y1,y2,'r*-')

plt.subplot(3,3,7)
plt.plot(y2,y3,color="black")

plt.subplot(3,3,8)
plt.plot(y3,y4,color="grey")

plt.subplot(3,3,9)
plt.plot(y4,y5,color="yellow")

```
